```python
# recursive func
def factorial(n):
    if(n==0 or n==1):
        return 1
    return n*factorial(n-1)

n=int(input("enter the number:  "))
print(f"the factorial of {n} is {factorial(n)}")
```

```python
# fac without using func
n=int(input("enter a number : "))
fac=1
for i in range(1,n+1):
    fac *=i

print(fac)
```
