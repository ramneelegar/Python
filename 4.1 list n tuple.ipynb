```python
# methods of list
a=[1,2,3,4,5]

a.append(6)
print(a)

print(sum(a))
a.reverse()
print(a)

a.insert(0,12)
print(a)

a.sort()
print(a)

a.index(4)
print(a)

a.pop()
print(a)

a.pop(1)
print(a)

a.count(1)
len(a)
print(a)

max(a)
print(a)

a.clear()
print(a)
```
