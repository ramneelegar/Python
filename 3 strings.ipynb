```python
# chapter 3.2 string slicing
# cutting strings in small pices
name="ramu"
short=name[0:3]
print(short)

a='raju neelegar'
print(a[0:4])

c="sapthagiri"
print(c[:5])

d='univercity'
print(d[0:])

e="bengaluru"
print(e[0:7:2])
# start:stop:steps
```

```python
# strind functions

# str lenght
# here k is the string for all function below
k="hodlur"
print(len(k))

#str start with or ends with
b=k.endswith('lur')
print(b)
c=k.startswith('ra')
print(c)

# capitalize 
r='raju'
d=r.capitalize()  #capitalize only first letter
print("caplital :",d)

# upper n lower case
e=k.lower()
print(e)
f=k.upper()
print(f)

# count the number of elements
g=k.count('h')
print(g)

# find the letter or word n return its index value
h=k.find('lur')
print(h)

# replace old word to new word
a='ramu'
j=a.replace('u',' neelegar')
print(j)

# split the str into list
i=k.split()
print(i)

# capitalize first char of each word
z="hii i m ram neelegar"
l=z.title()
print(l)
```
