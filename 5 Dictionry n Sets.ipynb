```python
def binary_search(arr, target):
    left, right = 0,len(arr) - 1
    
    while left <= right:
        mid = (left + right) // 2  # Find the middle index
        
        if arr[mid] == target:  # Found the element
            return mid
        elif arr[mid] < target:  # Search in the right half
            left = mid + 1
        else:  # Search in the left half
            right = mid - 1
    
    print("element not found")  # Element not found

# Example usage:
numbers = [2, 5, 7, 10, 15, 20, 25]
s=int(input("enter nu:"))
print(binary_search(numbers, s))  # Output: 3 (index of 10)
```
